```python
import pandas 
from sklearn import linear_model
import numpy as nm
import matplotlib.pyplot as mlt
```

```python
df = pandas.read_csv("Quote-Equity-CIPLA-EQ-08-09-2023-to-08-01-2024.csv")
columns_to_convert = ['OPEN ', 'HIGH ', 'LOW ', 'vwap ', 'VOLUME ', 'VALUE ','close ']

# Apply a lambda function to remove commas and convert to float for each specified column
df[columns_to_convert] = df[columns_to_convert].apply(lambda x: x.str.replace(',', '').astype(float))
X = df[['OPEN ', 'HIGH ','LOW ','vwap ','VOLUME ','VALUE ']]

y = df['close ']
print(X,y)
train_x=X.iloc[:64,:]
test_x=X.iloc[64:,:]
train_y=y.iloc[:64]
test_y=y.iloc[64:]

print( test_x,train_x)
```

```python
regr = linear_model.LinearRegression()
regr.fit(train_x, train_y)
```

```python
predictedstckp = regr.predict(test_x)
print(predictedstckp)
print(test_y)
```

```python
mlt.subplot(3,2, 1)
mlt.scatter(train_x["OPEN "], train_y)
mlt.plot(test_x["OPEN "],test_y,color='r')
mlt.plot(test_x["OPEN "],predictedstckp,color='b')
mlt.subplot(3,2, 2)
mlt.scatter(train_x["HIGH "], train_y)
mlt.plot(test_x["HIGH "],test_y,color='r')
mlt.plot(test_x["HIGH "],predictedstckp,color='b')
mlt.subplot(3,2, 3)
mlt.scatter(train_x["LOW "], train_y)
mlt.plot(test_x["LOW "],test_y,color='r')
mlt.plot(test_x["LOW "],predictedstckp,color='b')
mlt.subplot(3,2, 4)
mlt.scatter(train_x["vwap "], train_y)
mlt.plot(test_x["vwap "],test_y,color='r')
mlt.plot(test_x["vwap "],predictedstckp,color='b')
mlt.subplot(3,2, 5)
mlt.scatter(train_x["VALUE "], train_y)
mlt.plot(test_x["VALUE "],test_y,color='r')
mlt.plot(test_x["VALUE "],predictedstckp,color='b')
mlt.subplot(3,2, 6)
mlt.scatter(train_x["VOLUME "], train_y)
mlt.plot(test_x["VOLUME "],test_y,color='r')
mlt.plot(test_x["VOLUME "],predictedstckp,color='b')
mlt.show()
```
